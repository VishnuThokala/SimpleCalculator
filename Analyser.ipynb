```python
import pandas as pd

ds = pd.read_csv('whatsapp.csv')
List=[]
for i in range(len(ds)):
     List.append(ds.iloc[i,2])

s=set(List)
print(len(s),s)

di={}
for i in s:
  count = 0
  di[i]=0
  for j in range(len(ds)):
    if ds.iloc[j,2] == i:
      count=count+1
  di[i]=count;
  # print(i,"has sent ", count ,"messages")
print(di)
```
